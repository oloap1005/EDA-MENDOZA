```python
#LIBRARIES TO BE USED
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar 
```

#### LOADING THE FILE FOR THE DATASET TO BE ANALYZED

```python
# WE STORED THE DATASET AS D_STY THEN LOADED THE FILE
d_sty = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1') 
d_sty
```

#### **CHECKING THE DATA (GETTING TOTAL NON-NULL AND ALSO THE SUMMARY OF DATA  TYPES)

```python
## using the .info syntax to get datatype of each column
d_sty.info()
```

```python
# Check for missing values
missing= d_sty.isnull().sum()
print("\nMissing values in each column:\n", missing)
```

```python
## Getting the mean, median, and standard deviation of the streams column
d_sty ['streams'] = pd.to_numeric(d_sty ['streams'], errors='coerce')
mean_streams = d_sty['streams'].mean()
median_streams = d_sty['streams'].median()
std_dev_streams = d_sty['streams'].std()

print("Mean of streams:", mean_streams)
print("Median of streams:", median_streams)
print("Standard deviation of streams:", std_dev_streams)


```

```python
# Plot the distribution of 'released_year'
plt.figure(figsize=(12, 6))
plt.hist(d_sty['released_year'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Released Year')
plt.xlabel('Released Year')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of 'artist_count'
plt.figure(figsize=(12, 6))
plt.hist(d_sty['artist_count'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Artist Count')
plt.xlabel('Artist Count')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Calculate bounds
Q1 = d_sty['artist_count'].quantile(0.25)
Q3 = d_sty['artist_count'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# defining outliers
outliers = d_sty[(d_sty['artist_count'] < lower_bound) | (d_sty['artist_count'] > upper_bound)]

```

```python
# Plot data for the outliers
plt.figure(figsize=(10, 5))
plt.scatter(d_sty.index, d_sty['artist_count'], color='lightblue', label='Data')
plt.scatter(outliers.index, outliers['artist_count'], color='red', label='Outliers')

# Add quartile and boundary lines
for val, label, color in [(Q1, 'Q1', 'green'), (Q3, 'Q3', 'blue'),
                          (lower_bound, 'Lower Bound', 'orange'), (upper_bound, 'Upper Bound', 'purple')]:
    plt.axhline(val, linestyle='--', color=color, label=label)

# Labels and title
plt.title('Artist Count with Outliers')
plt.xlabel('Index')
plt.ylabel('Artist Count')
plt.legend()
plt.show()
```

```python
# Sort by 'streams' column in descending order and select the top 5 tracks
top_5_streamed_tracks = d_sty[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(5)
top_5_streamed_tracks
```

```python
# getting the top 5 artists 
top_5_artists = d_sty['artist(s)_name'].value_counts().head(5).reset_index()
top_5_artists.columns = ['artist_name', 'track_count']
top_5_artists
```

## TEMPORAL TRENDS

```python
# Count the number of tracks per year and sort by year
tracks_per_year = d_sty['released_year'].value_counts().sort_index()

# Plot the number of tracks released over time as a bar graph
plt.figure(figsize=(12, 6))
plt.bar(tracks_per_year.index, tracks_per_year.values, color='skyblue', edgecolor='black')
plt.title('Number of Tracks Released Over Time')
plt.xlabel('Released Year')
plt.ylabel('Number of Tracks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
## Count the number of releases per month and sort by month
releases_per_month = d_sty['released_month'].value_counts().sort_index()
```

```python
# Plot the number of releases per month
plt.figure(figsize=(12, 6))
sns.barplot(x=releases_per_month.index, y=releases_per_month.values, color='black')
plt.title('Number of Tracks Released Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

## Genre and Music Characteristics

```python
# Calculate the correlation matrix for streams and relevant musical attributes
correlation_matrix = d_sty[['streams', 'bpm', 'danceability_%', 'energy_%']].corr()
# Display the correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```

```python
# Identify which attributes have the strongest influence on streams
stream_correlation = correlation_matrix['streams'].drop('streams').abs().sort_values(ascending=False)
print("\nAttributes with the strongest influence on streams:\n", stream_correlation)
```

```python

# Calculate the correlation between 'danceability_%' and 'energy_%'
danceability_energy_corr = d_sty['danceability_%'].corr(d_sty['energy_%'])
danceability_energy_corr
```

```python
# visualizing the data to better explain it
# Plot danceability vs. energy
plt.figure(figsize=(8, 6))
plt.scatter(d_sty['danceability_%'], d_sty['energy_%'], alpha=0.5, color='black')
plt.title('Danceability vs. Energy')
plt.xlabel('Danceability (%)')
plt.ylabel('Energy (%)')
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation between valence and acousticness
valence_acousticness_corr = d_sty['valence_%'].corr(d_sty['acousticness_%'])
valence_acousticness_corr
```

```python
# Plot valence vs. acousticness
plt.figure(figsize=(8, 6))
plt.scatter(d_sty['valence_%'], d_sty['acousticness_%'], alpha=0.5, color='g')
plt.title('Valence vs. Acousticness')
plt.xlabel('Valence (%)')
plt.ylabel('Acousticness (%)')
plt.grid(True)
plt.show()
```

## Platform Popularity

```python
# Calculate basic statistics for the relevant columns
playlist_chart_stats = d_sty[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].describe()
playlist_chart_stats
```

```python
# Create a box plot to compare the distributions
plt.figure(figsize=(10, 6))
d_sty[['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists']].plot.box()
plt.title('Comparison of Track Counts in Playlists and Charts')
plt.ylabel('Number of Playlists/Charts')
plt.grid(True)
plt.show()
```

##  Advanced Analysis

```python
#  calculate for  the tracks for key and mode
key_mode=d_sty.groupby(['key', 'mode']).size().reset_index(name='Count')
key_mode
```

```python
#plotting the data that i got from the key and mode
plt.figure(figsize=(12, 8))
sns.barplot(data=key_mode, x='key', y='Count', hue='mode')
plt.title("Count of Songs by Key and Mode")
plt.xlabel("Key")
plt.ylabel("Count")
plt.legend(title="Mode")
plt.show()
```

```python
# grouping data for playlist and chart appearances.
artistplaylist_charts = d_sty.groupby('artist(s)_name')[[
    'in_spotify_playlists', 'in_apple_playlists', 'in_spotify_charts', 'in_apple_charts'
]].sum().sort_values(by='in_spotify_playlists', ascending=False)
artistplaylist_charts
```

```python
# Select top 5 artists for playlist and chart appearances
top_playlist_artists = artistplaylist_charts[['in_spotify_playlists', 'in_apple_playlists']].head(5)
top_chart_artists = artistplaylist_charts[['in_spotify_charts', 'in_apple_charts']].head(5)
top_playlist_artists
top_chart_artists
```

```python
# Plot for playlist appearances
plt.figure(figsize=(7, 6))
top_playlist_artists.plot(kind='bar', color=['blue', 'cyan'])
plt.title('Top 5 Artists by Playlist Appearances')
plt.xlabel('Artist')
plt.ylabel('Playlist Appearances')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Plot for chart appearances
plt.figure(figsize=(7, 6))
top_chart_artists.plot(kind='bar', color=['blue', 'cyan'])
plt.title('Top 5 Artists by Chart Appearances')
plt.xlabel('Artist')
plt.ylabel('Chart Appearances')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
